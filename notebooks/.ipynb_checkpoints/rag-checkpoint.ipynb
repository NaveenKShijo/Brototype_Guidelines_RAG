<h2>Building the retrieval layer (LlamaIndex)</h2>

```python
from llama_index.core import SimpleDirectoryReader

documents = SimpleDirectoryReader("data").load_data()
# SimpleDirectoryReader() looks inside the folder named 'data'. It reads all the supported files inside it. Then the files are loaded 
# and each file is converted into LlamaIndex document objects. Then we get a list of documents
```

```python
documents
```

```python
# setting up embedding model - Using a local HuggingFace embedding model

from llama_index.embeddings.huggingface import HuggingFaceEmbedding

embed_model = HuggingFaceEmbedding(
    model_name = "BAAI/bge-small-en-v1.5"
)

# This is just a warning. Your code will still work.
#     Reason:
#     Windows without Developer Mode cannot create symbolic links.
# You can ignore it.
```

<h2>Connect Chroma vector database</h2>

```python
import chromadb
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.core import StorageContext
```

```python
# Creating database

chroma_client = chromadb.Client()
# This creates a connection to a chroma vector database instance which is created in RAM. Data exists only while python 
# program running. When the program stops, the database disappears

# to store it on SSD(persistent storage)
# from chromadb.config import Settings
# import chromadb

# chroma_client = chromadb.Client(
#     Settings(
#         persist_directory="./chroma_db"
#     )
# )

collection = chroma_client.get_or_create_collection("rag_db")
# This creates a collection (like a table) inside the vector database with the name 'rag_db'

vector_store = ChromaVectorStore(chroma_collection = collection)
# LlamaIndex does not store vectors. Instead it uses external vector databases. This line creates a 
# bridge adaptor(ChromaVectorStore) between LlamaIndex and ChromaDB

storage_context = StorageContext.from_defaults(
    vector_store = vector_store
)
# This tells LlamaIndex where to store it's data. StorageContext is a container that manages storage backends.
```

<h2>Create the index</h2>

This step performs 
    
    documents -> chunking -> embedding -> vector storage

```python
from llama_index.core import VectorStoreIndex

index = VectorStoreIndex.from_documents(
    documents,
    storage_context = storage_context,
    embed_model = embed_model
)

# Now the knowledge base is ready
```

```python
retriever = index.as_retriever(similarity_top_k = 5)                               
```

<h2>Connect Retrieval to LangChain</h2>

LangChain handles this chain:
    
    User question
    ↓
    LangChain pipeline
    ↓
    Call LlamaIndex retriever
    ↓
    Insert retrieved context into prompt
    ↓
    Send to Gemini

```python
def retrieve_context(query):
    nodes = retriever.retrieve(query)
    context = "\n\n".join([node.text for node in nodes])
    return context
```

```python
query = "What are the rules to be followed at the Brototype hub regarding wearables"
nodes = retriever.retrieve(query)
len(nodes)
```

```python
nodes[1].text
```

<h2>Prompt construction</h2>

```python
def generate_answer(question):
    context = retrieve_context(question)
    prompt = f"""
        Use this context to answer the question below.

        Context:
        {context}

        Question:
        {question}        
    """
    response = llm.invoke(prompt)
    return response.content
```

<h2>Running the RAG system</h2>

```python
while True:
    query = input("Ask: ")
    answer = generate_answer(query)
    print("\nAnswer: ", answer) 
```
